```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

%matplotlib inline
```

## The following table shows the demand and supply schedules of the market for electric vehicles.

```python
title_text = 'Market for Electric Vehicles'
data =  [
            [ 'Price\n($/EV unit)', 'Initial Quantity of\nElectric Vehicles Supplied',
             'Initial Quantity of\nElectric Vehicles Demanded',
             'New Quantity of\nElectric Vehicles Demanded\nWhen Demand Decreases', 
             'New Quantity of\nElectric Vehicles Supplied\nWhen Supply Increases'],
            [ '$91,000.00', '100,000', '25,000', '-', '-'],
            ['$84,000.00',   '90,000', '30,000', '-', '-'],
            ['$77,000.00',  '80,000',  '35,000', '-', '-'],
            ['$70,000.00',  '70,000',  '40,000', '-', '-'],
            ['$63,000.00',  '60,000',  '45,000', '-', '-'],
            ['$56,000.00',  '50,000',  '50,000', '-', '-'],
            ['$49,000.00',  '40,000',  '55,000', '-', '-'],
            ['$42,000.00',  '30,000',  '60,000', '-', '-'],
            ['$35,000.00',  '20,000',  '65,000', '-', '-'],
            ['$28,000.00',  '10,000',  '70,000', '-', '-'],
        ]
# Pop the header from the data array
column_headers = data.pop(0)

    
# Get some lists of color specs for column headers
ccolors = plt.cm.BuPu(np.full(len(column_headers), 0.1))


fig_background_color = 'azure'
fig_border = 'steelblue'

# Create the figure
plt.figure(linewidth=2,
           figsize=(20,150),
           edgecolor=fig_border,
           facecolor=fig_background_color)

# Add a table at the bottom of the axes
the_table = plt.table(cellText=data,
                      colColours=ccolors,
                      colLabels=column_headers,
                      loc='center')

# Set fontsize for table contents
the_table.auto_set_font_size(False)
the_table.set_fontsize(150)
# Make the rows taller
the_table.scale(10, 50)

# Hide axes
ax = plt.gca() 
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

# Set title
ax.set_title(title_text, 
             fontweight="bold",
             fontsize=250)

# Hide axes border
plt.box(on=None)

# Create image
fig = plt.gcf()
plt.savefig('pyplot-table-demo.png',
            edgecolor=fig.get_edgecolor(),
            facecolor=fig.get_facecolor(),
            dpi=150)
```

1) Draw the original market equilibrium graph. Label the demand curve D1 and the supply 
curve S1. Please mark the equilibrium with a dot. Please give a title to your graph and make sure to label the x-axis and the y-axis.

```python
price = [28000, 35000, 42000, 49000,
        56000, 63000, 70000, 77000,
        84000, 91000]

supply1 = [10000, 20000, 30000, 40000,
           50000, 60000, 70000, 80000,
           90000, 100000]
demand1 = [70000, 65000, 60000, 55000,
           50000, 45000, 40000, 35000,
           30000, 25000]

plt.figure(figsize=(12,12))

# Plot supply curve and demand curve
plt.plot(supply1, price, 'red', demand1, price, 'dodgerblue',
         linewidth=5.0)

# Set the axis limits
plt.axis((0,120000,0,98000))
plt.xticks(np.arange(0,120000,10000))
plt.yticks(np.arange(0,98000,7000))

# Labels
plt.title('Market for Electric Vehicles', fontsize=17)
plt.xlabel('Quantity (EV units)', fontsize=16)
plt.ylabel('Price (dollars per EV unit)', fontsize=16)

# Set equilibrium point
plt.plot(50000, 56000, 'ro', color='black', markersize=20)

# Annotations
plt.annotate('D1', xy=(23000, 93000), size=13)
plt.annotate('S1', xy=(101000, 93000), size=13)

plt.grid()
plt.show()
```

3) Draw the graph you made for the first problem. On this graph, draw the new demand curve and the new supply curve. Label the new demand curve D2 and the new supply curve S2. Please mark the new equilibrium with a dot. Please give a title to your graph and make sure to label the x-axis and the y-axis.

```python
price = [28000, 35000, 42000, 49000,
        56000, 63000, 70000, 77000,
        84000, 91000]

supply1 = [10000, 20000, 30000, 40000,
           50000, 60000, 70000, 80000,
           90000, 100000]
demand1 = [70000, 65000, 60000, 55000,
           50000, 45000, 40000, 35000,
           30000, 25000]

supply2 = [20000, 30000, 40000,
          50000, 60000, 70000, 80000,
          90000, 100000, 110000]
demand2 = [50000, 45000, 40000, 35000,
           30000, 25000, 20000, 15000,
           10000, 5000]

plt.figure(figsize=(12,12))

# Plot supply curve and demand curve
plt.plot(supply1, price, 'red', demand1, price, 'dodgerblue',
         supply2, price, 'red', demand2, price, 'dodgerblue',
         linewidth=5.0)

# Set the axis limits
plt.axis((0,120000,0,98000))
plt.xticks(np.arange(0,120000,10000))
plt.yticks(np.arange(0,98000,7000))

# Labels
plt.title('Market for Electric Vehicles', fontsize=17)
plt.xlabel('Quantity (EV units)', fontsize=16)
plt.ylabel('Price (dollars per EV unit)', fontsize=16)

# Set equilibrium point
plt.plot(40000, 42000, 'ro', color='black', markersize=20)

# Annotations
plt.annotate('D1', xy=(23000, 93000), size=13)
plt.annotate('S1', xy=(101000, 93000), size=13)
plt.annotate('D2', xy=(3000, 93000), size=13)
plt.annotate('S2', xy=(111000, 93000), size=13)

plt.grid()
plt.show()
```
